```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Load the Data

basket = pd.read_csv (r'Ahmed Project\ML AHMED\groceries.csv')

basket.head()
```

```python
basket.info()
```

```python
basket.describe().T
```

```python
# use encoding to se the data to be used for assoication 

basket = pd.get_dummies(basket)
basket.head()
```

```python
# use apriori and fix support to .03 to see how many itemsets are there with this condition.

apriori(basket, min_support=0.03)
```

```python
# fix the name of these item sets 
apriori(basket, min_support=0.03, use_colnames=True)
```

```python
# use apirori to find the ietmsets with support of .003 and fix the number of these utemsets 

itemsets = apriori(basket, min_support=0.003, use_colnames=True)

itemsets['length'] = itemsets['itemsets'].apply(lambda x: len(x))
itemsets
```

```python
itemsets['itemsets'].value_counts().sort_values(ascending=False).head(10)
```

```python
# filter by the itemsts with the length of 3 

itemsets[itemsets['length'] >= 3]
```

```python
# use the association rule to find items combination with the higest support

rule = association_rules(itemsets, metric="confidence", min_threshold=0.6)
rule.head()
```

```python
# use the association rule to find items combination with the higest lift

rule = association_rules(itemsets, metric="lift", min_threshold=0.6)
rule.head()
```

```python
# filter the itemsets when lift is more tha 5 and confidence is more than 06
rule[(rule['lift'] >= 5) & (rule['confidence']>= 0.6)]
```

```python
rule[(rule['lift'] >= 5) & (rule['confidence']>= 0.6)].rank()
```
